```python
from ultralytics import YOLO

# 1. Load the lightweight YOLOv8 segmentation model
model = YOLO("yolov8n-seg.pt")

# 2. Test the training loop (replace the path with the location of your data.yaml)
results = model.train(
    data='C:/Code/Year_2/Phase3/Intro_to_DS_AI/OCR-Receipt-recognition-and-Tracking/data.yaml', 
    epochs=50,        # Just 3 epochs for a fast test run
    imgsz=640,       # Standard image size
    batch=16,        # Safe batch size
    project="receipt_project",
    name="test_training"
)
```

```python
import cv2
import matplotlib.pyplot as plt
from ultralytics import YOLO
import os

# 1. Load YOUR newly trained model
my_custom_model = YOLO("yolov8n-seg.pt")

# 2. Pick a test image - find the first available test image
test_images_dir = "Receipt/test/images"
test_images = [f for f in os.listdir(test_images_dir) if f.lower().endswith(('.jpg', '.png', '.jpeg'))]

if test_images:
    test_image_path = os.path.join(test_images_dir, test_images[0])
    print(f"Using test image: {test_image_path}")
    
    # 3. Run prediction
    results = my_custom_model.predict(source=test_image_path, save=True, show=False)
    
    # 4. Display the result in the notebook
    # YOLO saves the annotated image in the 'runs/predict' folder by default
    output_img = cv2.imread(results[0].save_dir + "/" + test_images[0])
    if output_img is not None:
        output_img = cv2.cvtColor(output_img, cv2.COLOR_BGR2RGB)
        
        plt.figure(figsize=(10,10))
        plt.imshow(output_img)
        plt.axis('off')
        plt.show()
    else:
        print("Could not read output image")
else:
    print("No test images found in Receipt/test/images/")
```
